```python
import timeit
from imutils import paths
import cv2
import numpy as np
import features
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
resize = 0
args = {'input': './data/building', 'output': './result2', 'resize': 0}
image_path = list(paths.list_images(args["input"]))
list_image = []
for _, j in enumerate(image_path):
    print("loadImages _for_ image_path : ", _, j)
    image = cv2.imread(j)
    if resize == 1:
        image = cv2.resize(
            image, (int(image.shape[1] / 4), int(image.shape[0] / 4))
        )
    list_image.append(image)
```

```python
#extract keypoints and descriptors using sift
k0,f0=features.findAndDescribeFeatures(list_image[0],opt='SIFT')
k1,f1=features.findAndDescribeFeatures(list_image[1],opt='SIFT')
```

```python
def convertResult(img):
    '''Because of your images which were loaded by opencv, 
    in order to display the correct output with matplotlib, 
    you need to reduce the range of your floating point image from [0,255] to [0,1] 
    and converting the image from BGR to RGB:'''
    img = np.array(img,dtype=float)/float(255)
    img = img[:,:,::-1]
    return img
```

```python
#draw keypoints
img0_kp=features.drawKeypoints(list_image[0],k0)
img1_kp=features.drawKeypoints(list_image[1],k1)

plt_img = np.concatenate((img0_kp, img1_kp), axis=1)
plt.figure(figsize=(15,15))
plt.imshow(convertResult(plt_img))
```

```python
imgRGB = cv2.cvtColor(list_image[0], cv2.COLOR_BGR2RGB)  # BGR(OpenCV) -> RGB(PyQt5)
imgRGB1 = cv2.cvtColor(list_image[1], cv2.COLOR_BGR2RGB)  # BGR(OpenCV) -> RGB(PyQt5)
plt.subplot(221)
plt.imshow(imgRGB)
plt.subplot(222)
plt.imshow(imgRGB1)
```
